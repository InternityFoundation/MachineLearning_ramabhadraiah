```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(1,10)
x
```

```python
y1=2*x
y2=x^2
plt.plot(x,y1,color="green",linewidth=2,linestyle=":")
plt.plot(x,y2,color="green",linewidth=2,linestyle=":")
plt.xlabel("x-ax")
plt.ylabel("y-ax")
plt.title("plot")
plt.show()
```
